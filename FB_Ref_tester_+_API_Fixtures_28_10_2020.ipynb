```
home_table1920.rename(columns={'ï»¿Squad': 'team'}, inplace=True)
home_table1920 = home_table1920.replace('Wolverhampton Wanderers','Wolverhampton')
home_table1920 = home_table1920.replace('Leeds United', 'Leeds Utd')
home_table1920 = home_table1920.replace('West Ham', 'West Ham Utd')
home_table1920
```

```
homebothyears = home_table.merge(home_table1920, on="team", left_index=True, right_index=True)
homebothyears['MP']=homebothyears['MP_x']+homebothyears['MP_y']
homebothyears['GF']=homebothyears['GF_x']+homebothyears['GF_y']
homebothyears['GA']=homebothyears['GA_x']+homebothyears['GA_y']
homebothyears['H Avg GS'] = homebothyears ['GF'] / homebothyears ['MP']
homebothyears ['H Avg GA'] = homebothyears ['GA'] / homebothyears ['MP']
homebothyears ['Home LA GS'] = homebothyears['H Avg GS'].mean()
homebothyears ['Home LA GA'] = homebothyears['H Avg GA'].mean()
homebothyears ['H AS'] = homebothyears['H Avg GS']/homebothyears ['Home LA GS']
homebothyears ['H DS'] = homebothyears['H Avg GA']/homebothyears ['Home LA GA']
home_table_reduced = homebothyears[['team', 'H AS', 'H DS','Home LA GS']]
home_table_reduced
```

```
away_table1920.rename(columns={'ï»¿Squad': 'team'}, inplace=True)
away_table1920 = away_table1920.replace('Wolverhampton Wanderers','Wolverhampton')
away_table1920 = away_table1920.replace('Leeds United', 'Leeds Utd')
away_table1920 = away_table1920.replace('West Ham', 'West Ham Utd')
away_table1920
```

```
awaybothyears = away_table.merge(away_table1920, on="team", left_index=True, right_index=True)
awaybothyears['MP']=awaybothyears['MP_x']+awaybothyears['MP_y']
awaybothyears['GF']=awaybothyears['GF_x']+awaybothyears['GF_y']
awaybothyears['GA']=awaybothyears['GA_x']+awaybothyears['GA_y']
awaybothyears['A Avg GS'] = awaybothyears ['GF'] / awaybothyears ['MP']
awaybothyears ['A Avg GA'] = awaybothyears ['GA'] / awaybothyears ['MP']
awaybothyears ['Away LA GS'] = awaybothyears['A Avg GS'].mean()
awaybothyears ['Away LA GA'] = awaybothyears['A Avg GA'].mean()
awaybothyears ['A AS'] = awaybothyears['A Avg GS']/awaybothyears ['Away LA GS']
awaybothyears ['A DS'] = awaybothyears['A Avg GA']/awaybothyears ['Away LA GA']
away_table_reduced = awaybothyears[['team', 'A AS', 'A DS','Away LA GS']]
away_table_reduced
```

```
combined_table = pd.merge(home_table_reduced, away_table_reduced, on='team')
combined_table.sort_values('team')
combined_table
```

```
from urllib.request import urlopen
from json import loads
from csv import writer


def team_converter(team_id):
    '''Converts a team's ID to their actual name'''
    team_map = {
        1: "Arsenal",
        2: "Aston Villa",
        3: "Brighton",
        4: "Burnley",
        5: "Chelsea",
        6: "Crystal Palace",
        7: "Everton",
        8: "Fulham",
        9: "Leicester City",
        10: "Leeds Utd",
        11: "Liverpool",
        12: "Manchester City",
        13: "Manchester Utd",
        14: "Newcastle Utd",
        15: "Sheffield Utd",
        16: "Southampton",
        17: "Tottenham",
        18: "West Brom",
        19: "West Ham Utd",
        20: "Wolves",
        None: None
    }
    return team_map[team_id]


def position_converter(position):
    '''Converts a player's element_type to their actual position'''
    position_map = {
        1: "Goalkeeper",
        2: "Defender",
        3: "Midfielder",
        4: "Forward"
    }
    return position_map[position]


def main():
    all_data = loads(urlopen("https://fantasy.premierleague.com/api/fixtures/").read())
    fixtures = all_data

    important_data = [
        [
            x["event"],
            team_converter(x["team_h"]),
            team_converter(x["team_a"]),
            x["kickoff_time"],
            
        ]
        for x in fixtures
    ]

    with open("/fixtures_data.csv", "w", encoding="utf-8", newline="") as out:
        headers = ["event","team_h","team_a","kickoff_time"]
        w = writer(out)
        w.writerow(headers)
        w.writerows(important_data)

if __name__ == "__main__":
    main()
```
